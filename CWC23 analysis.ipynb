## **1. Team Performance Analysis**

```python
# Explore team-wise performance metrics
team_performance = data.groupby('team').agg({
    'runs': 'sum',
    'wkts': 'sum',
    'bb_bf': 'sum',
    'runs_per_ball': 'mean',
    'econ': 'mean',
    'inns': 'sum'
}).reset_index()

print(team_performance)
```

```python
# Identify top-performing teams and their strengths
top_teams_runs = team_performance.nlargest(5, 'runs')
top_teams_wkts = team_performance.nlargest(5, 'wkts')

print("Top 5 Teams by Runs:")
print(top_teams_runs)

print("\nTop 5 Teams by Wickets:")
print(top_teams_wkts)
```

```python
# Visualize runs scored by each team
plt.figure(figsize=(12, 6))
plt.bar(team_performance['team'], team_performance['runs'])
plt.xlabel('Team')
plt.ylabel('Total Runs Scored')
plt.title('Total Runs Scored by Each Team')
plt.xticks(rotation=45)
plt.show()
```

## **2. Player Performance Analysis**

```python
# Evaluate individual player statistics for both batting and bowling
player_performance = data.groupby(['team', 'player', 'bat_or_bowl']).agg({
    'runs': 'sum',
    'wkts': 'sum',
    'bb_bf': 'sum',
    'runs_per_ball': 'mean',
    'econ': 'mean',
    'inns': 'sum'
}).reset_index()

print(player_performance.head())
```

```python
# Identify leading run-scorers and wicket-takers
leading_run_scorers = player_performance[player_performance['bat_or_bowl'] == 'bat'].nlargest(10, 'runs')
leading_wicket_takers = player_performance[player_performance['bat_or_bowl'] == 'bowl'].nlargest(10, 'wkts')

print("Leading Run-scorers:")
print(leading_run_scorers)

print("\nLeading Wicket-takers:")
print(leading_wicket_takers)
```

## **3. Opposition and Ground Analysis**

```python
# Investigate how teams and players perform against different oppositions
opposition_analysis = data.groupby(['team', 'opposition']).agg({
    'runs': 'sum',
    'wkts': 'sum',
    'bb_bf': 'sum'
}).reset_index()

print(opposition_analysis)
```

## **4. Temporal Analysis**

```python
# Study performance trends over time
temporal_analysis = data.groupby('start_date').agg({
    'runs': 'sum',
    'wkts': 'sum',
    'bb_bf': 'sum'
}).reset_index()

print(temporal_analysis.head())
```

```python
# Visualize performance trends over time
plt.figure(figsize=(12, 6))
plt.plot(temporal_analysis['start_date'], temporal_analysis['runs'], label='Total Runs')
plt.plot(temporal_analysis['start_date'], temporal_analysis['wkts'], label='Total Wickets')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Performance Trends Over Time')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
